```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

# Generate random data for three Gaussian blobs in 3D space
n_samples = 30



data1 = np.clip(np.random.multivariate_normal([4, 1], cov=[[0.15, 0], [0, 0.2]], size=n_samples), 0, 5)
data2 = np.clip(np.random.multivariate_normal([0.5, 0.5], cov=[[1, 0], [0, 1]], size=n_samples), 0, 5)
data3 = np.clip(np.random.multivariate_normal([4.2, 3.8], cov=[[0.11, 0], [0, 0.5]], size=n_samples), 0, 5)
data4 = np.clip(np.random.multivariate_normal([2.5, 4.8], cov=[[1, 0], [0, 0.1]], size=n_samples), 0, 5)


data = np.concatenate((data1, data2, data3, data4), axis=0)
kmeans = KMeans(n_clusters=4).fit(data)


# Create a 3D scatter plot
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111)
#ax.scatter(data1[:,0], data1[:, 1], c='r', marker='o', label='Blob 1', s=100)
#ax.scatter(data2[:,0], data2[:, 1], c='g', marker='o', label='Blob 2', s=100)
#ax.scatter(data3[:,0], data3[:, 1], c='b', marker='o', label='Blob 2', s=100)
#ax.scatter(data4[:,0], data4[:, 1], c='y', marker='o', label='Blob 2', s=100)

ax.scatter(data[:,0], data[:, 1], c=kmeans.labels_, marker='o', label="", s=150, cmap="RdYlGn")

# Set axis labels
ax.set_xlabel('Animation Movie Rating')
ax.set_ylabel('Action Movie Rating')
ax.grid("on")
ax.legend()
# Add a legend
# ax.legend()

# Show the plot

plt.title("Movie Rating Clustering")
plt.show()
```
