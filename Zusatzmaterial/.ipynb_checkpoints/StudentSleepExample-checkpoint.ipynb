```python
import random

# Generate an expanded dataset with 100 data points
random.seed(0)  # Set a seed for reproducibility
age = [random.randint(30, 70) for _ in range(100)]
cholesterol_level = [random.randint(150, 250) for _ in range(100)]
condition = ["Yes" if random.random() < 0.5 else "No" for _ in range(100)]

# Create a scatter plot
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(age, cholesterol_level, c=[1 if c == "Yes" else 0 for c in condition], cmap='coolwarm', marker='o')
plt.xlabel("Age (years)")
plt.ylabel("Cholesterol Level (mg/dL)")
plt.title("Scatter Plot of Age vs. Cholesterol Level")
plt.colorbar(label="Condition (1: Yes, 0: No)")

# Show the plot
plt.show()
```
